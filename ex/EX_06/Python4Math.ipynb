## Datatypes 

Python has a number of fundamental types such as list/arrays, tuples, and dictionaries.

```python
# list
list_of_numbers = [0,1,2,3]
list_of_words = ['red', 'blue', 'green']

print(list_of_numbers)
print(list_of_words)
list_of_words[0]  # list are indexed starting at zero
```

```python
# tuples
tuple_of_stuff=(1,2,'green') 

print(tuple_of_stuff)
tuple_of_stuff[-1] # you can also index tuples, i.e. starting at the last element
```

```python
# tuples are immutable, meaning that they cannot be changed
tuple_of_stuff[-1] = 'red'
```

```python
# dictionaries are a fundamental datatype in Python
dictionary = {'birds':['robin', 'bluejay', 'falcon'],
              'monsters':['Grendel', 'Hircocervus', 'Manticore']}
print(dictionary.keys())
print(dictionary.values())
```

```python
dictionary['birds'] # we can get values by key
```

```python
dictionary['monsters'][0] # we can index a values
```

```python
dictionary[0] # we cannot index a key
```

Python has a number of builtin and external libraries.

When we import a library we map its name-space into the Python's standard name-space.
Don't worry about this too much, but we might not want to overwrite our standard name-space, so we'll given the library name-space an short alias, such as np, sym, or plt (these are the common alias used by convention.)

## Plotting

Parameteric plotting means that we create a list of values and apply our funciton at each value.

```python
x = np.arange(0, 2 * np.pi, 0.1)
x
```

```python
plt.plot(x, x)
```

Plotting a symbolic function is a bit more abstract.
We'll use what is called a CAS (Computer Algebra System).
This is exactly how a graphing calculator plots graphs.
It's important to understand that, only parametric plotting is possible on a computer.
It's also worth noting that not all (most) problems don't have analytical solutions and must be solved using numerical techniques.--We'll ignore this for now, but in Data Science approximation is often the best we can do. 

```python
x = sym.symbols('x') # symbols are an abstract datatype used by sympy
sym.plot( x, sym.sin(x), sym.cos(x), (x, -sym.pi, sym.pi))
```

Before we move on let's plot some arbitrary polynomial function.

```python
x = sym.symbols('x') 
f= (x**4 - 15*x**3 + 80*x**2 - 180*x + 144)/10
f # Sympy has 'pretty-printing' for Jupyter Notebook and Lab
```

```python
# Instead of the print() method we can use display() for pretty-printing
display(f)
```

```python
# Latex markdown can be used to create rich typesetting
from sympy import latex  # import a single method from a name-space for ease
from IPython.display import Markdown

display( Markdown(
    "We can get super fancy using the $\LaTeX$ markdown language!!!\n\n"
    + "Here we use a multiline string to generate our markdown.\n\n" 
    + '$'
    + 'f='
    + f'{latex(f)}'
    + '$'
))
sym.plot(f, (x,1,6.5)); # semicolons suppress output
```

```python
import sympy as sym
from sympy import solve

x = sym.symbols('x') 
f= (x**4 - 15*x**3 + 80*x**2 - 180*x + 144)/10
zero_points = solve(f, x)

args = [ {'args':[p,0,'go']} for p in zero_points ] # list comprehension of a list of dictionaries (complicated yet powerful) 
sym.plot(f, (x,1,6.5), markers=args);
```
